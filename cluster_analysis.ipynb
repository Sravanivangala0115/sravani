```python
data['Annual Income (k$)'].mean(), data['Spending Score (1-100)'].mean()
```

```python
X.values
```

```python
wss=[]
for i in range(1,21):
    kmeans_1=KMeans(n_clusters=i,max_iter=300,random_state=1234)
    kmeans_1.fit(X)
    wss.append(round(kmeans_1.inertia_,2))
```

```python
wss
```

```python
plt.scatter(range(1,21),wss)
plt.plot(range(1,21),wss)
plt.title("The Elbow method")
plt.xlabel("clusters")
plt.ylabel("wss")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,
                max_iter=300,
                random_state=1234)
y_clusters=kmeans.fit_predict(X)
y_clusters
```

```python
data['Cluster_group']=y_clusters
```

```python
data.head()
```

```python
data['Cluster_group'].value_counts()
```

```python
X=data.iloc[:,[3,4]].values
plt.scatter(X[y_clusters==0,0],X[y_clusters==0,1],s=100,c='red',label="cluster")
plt.scatter(X[y_clusters==1,0],X[y_clusters==1,1],s=100,c='blue',label="cluster")
plt.scatter(X[y_clusters==2,0],X[y_clusters==2,1],s=100,c='green',label="cluster")
plt.scatter(X[y_clusters==3,0],X[y_clusters==3,1],s=100,c='black',label="cluster")
plt.scatter(X[y_clusters==4,0],X[y_clusters==4,1],s=100,c='orange',label="cluster")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',
            label='Centroids')
plt.title("Clusters using k-means")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
